````python
import os

os.environ['OPENAI_BASE_URL'] = "https://llm-api.cyverse.ai"
os.environ['OPENAI_API_KEY'] = "..."

model_name = "Meta-Llama-3.1-70B-Instruct-quantized"
````

````python
import os

from llama_index.llms.openllm import OpenLLM
from llama_index.core.llms import ChatMessage


llm = OpenLLM(
    model=model_name, api_base=os.environ['OPENAI_BASE_URL'], api_key=os.environ['OPENAI_API_KEY']
)


completion_response = llm.complete("To infinity, and")
print(completion_response)
````

````python
from pprint import pprint

response = llm.chat(
    [
        ChatMessage(
            role="system", content="You are acting as Ernest Hemmingway."
        ),
        ChatMessage(role="user", content="Hi there!"),
        ChatMessage(role="assistant", content="Yes?"),
        ChatMessage(role="user", content="What is the meaning of life?"),
    ]
)

pprint(response.message.content)
````

````python
# Download a very long document: Harry Potter Vol 1
import requests  # the lib that handles the url stuff

target_url = """https://raw.githubusercontent.com/amephraim/nlp/refs/heads/master/texts/J.%20K.%20Rowling%20-%20Harry%20Potter%201%20-%20Sorcerer's%20Stone.txt"""
response = requests.get(target_url) # it's a file like object and works just like a file
text = response.text # files are iterable
print(text)
````

````python
from pprint import pprint
from llama_index.llms.openllm import OpenLLM
from llama_index.llms.ollama import Ollama
# llm = Ollama(model="llama3.1:latest")

llm = OpenLLM(
    model=model_name, api_base=os.environ['OPENAI_BASE_URL'], api_key=os.environ['OPENAI_API_KEY']
)

response = llm.chat(
    [
        ChatMessage(
            role="system", content="You are a copy editor helping in the editorial process of publishing books"
        ),
        ChatMessage(role="user", content="""
					Book contents:
					```{contents}```
					
					Summarize the book into three paragraphs.""".strip().format(contents=text[:335000])),
    ]
)

pprint(response)
````
